```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cross_decomposition import PLSCanonical, PLSRegression, CCA
```

```python
data = load_breast_cancer()
X = data.data
Y = data.target

plsca = PLSCanonical(n_components=1)
plsca.fit(X, Y)
X_plsca, Y_plsca = plsca.transform(X, Y)

pls2 = PLSRegression(n_components=1)  
pls2.fit(X, Y)
X_pls2, Y_pls2 = pls2.transform(X, Y)

cca = CCA(n_components=1) 
cca.fit(X, Y)
X_cca, Y_cca = cca.transform(X, Y)

plt.figure(figsize=(12, 8))
plt.subplot(221)
plt.scatter(X_plsca[:, 0], Y_plsca[:, 0], c=Y, cmap='viridis')
plt.xlabel("X PLSCanonical")
plt.ylabel("Y PLSCanonical")
plt.title("PLSCanonical")

plt.subplot(222)
plt.scatter(X_pls2[:, 0], Y_pls2[:, 0], c=Y, cmap='viridis')
plt.xlabel("X PLS2")
plt.ylabel("Y PLS2")
plt.title("PLSRegression (PLS2)")

plt.subplot(223)
plt.scatter(X_cca[:, 0], Y_cca[:, 0], c=Y, cmap='viridis')
plt.xlabel("X CCA")
plt.ylabel("Y CCA")
plt.title("CCA")

plt.tight_layout()
plt.show()
```
